```python
#Clean up monetary values
monetary = ["Value", "Wage", "Release Clause"]
#             €M   ,  €K   ,   €M

def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

for col in monetary:
    df[col] = df[col].str.strip('€')
    df[col] = df[col].apply(value_to_float)


#df[wage] = df1["Wage"].replace('[\€, ]', '', regex=True).apply(value_to_float) #replaces euro, comma and whitespace with ''
#df["Value"].replace(['Z', 'M'],['',''])
print(df[monetary])
```
